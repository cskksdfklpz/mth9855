## Problem 5.4

In APT model, the covariance matrix of the return is 

$$
\Sigma = \mathrm{Cov}(R) = XFX^T+D
$$

where $F = \mathrm{Cov}(f)$ is a s.p.d. matrix and $D = \mathrm{diag}(\sigma_1^2,\dots,\sigma_n^2)$ is a diagonal matrix. So we know that $F$ has a Cholesky decomposition $F = LL^T$ which means

$$
XFX^T = XLL^TX^T = (XL) (XL)^T
$$

Then we can apply Woodbury matrix identity

$$
\begin{aligned}
    \Sigma^{-1} &= \left[D + (XL)(XL)^T\right]^{-1}\\
                &= D^{-1} - D^{-1} (XL) \left[I_p+(XL)^T D^{-1}(XL)\right]^{-1} (XL)^T D^{-1}\\
\end{aligned}
$$

## Problem 5.5

1. (1) = (2)

Consider the optimization problem

$$
\begin{aligned}
    \min_b & & || Y-Xb ||^2 + \delta || b ||^2
\end{aligned}
$$

if the solution of $\arg\min_b || Y-Xb ||$ is not unique, the optimization above will give us the least-norm solution. And we can solve that problem analytically

$$
b = (X^TX+\delta I_p)^{-1} X^T Y
$$

So if $\delta\to 0$, $\arg\min_b || Y-Xb ||$ and $\arg\min_b || Y-Xb ||^2 + \delta || b ||^2$ will have the same optimal value, so we know that

$$
b_{\text{least norm}} = \lim_{\delta \to 0^+} (X^TX+\delta I_p)^{-1} X^T Y
$$

2. (2) = $X^+Y$

Notice that for $P = XX^+$ we have $PX = X, P = P^T$, so

$$
\begin{aligned}
    X^T(XX^+Y - Y) &= X^T (PY - Y)\\
                   &= X^T P^T Y - X^T Y\\
                   &= (PX)^T Y - X^T Y\\
                   &= X^T Y - X^T Y\\
                   &= 0\\
\end{aligned}
$$

which means for $\forall b$

$$
\begin{aligned}
    ||Xb-Y||^2 &= ||XX^+Y - Y||^2 + ||X(b - X^+Y)||^2 + 2(X(b-X^+Y)^T(XX^+Y-Y))\\
               &= ||XX^+Y - Y||^2 + ||X(b - X^+Y)||^2 + 2((b-X^+Y)^TX^T(XX^+Y-Y))\\
               &= ||XX^+Y - Y||^2 + ||X(b - X^+Y)||^2\\
               &\ge ||XX^+Y - Y||^2
\end{aligned}
$$
